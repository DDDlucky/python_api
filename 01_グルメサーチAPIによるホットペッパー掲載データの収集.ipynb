```python
import requests
```

```python
URL = 'https://webservice.recruit.co.jp/hotpepper/gourmet/v1/'
```

```python
API_KEY = '85b9f52920308a09'
```

```python
params = {
    'key' : API_KEY,
    'keyword' : "沖縄",
    'count': 100,
    'format' : 'json'
}
```

```python
res = requests.get(URL, params)
```

```python
res.status_code
```

```python
result = res.json()
```

```python
items = result['results']['shop']
```

```python
len(items)
```
